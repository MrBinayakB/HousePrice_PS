```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
from sklearn.metrics import accuracy_score
sns.set(style="white",color_codes=True)
```

```python
HouseDS=pd.read_csv("HousePS.csv")
HouseDS.head()
```

```python
HouseDS.info()
```

```python
#Filtering the Address
Filter_Address= [item.split(",")[1].strip() for sublist in [HouseDS['Address']] for item in sublist]
print(len(Filter_Address))
        
```

```python
#Adding Filter Address as New Column in Dataset
HouseDS=pd.DataFrame(HouseDS)
New_Address=pd.DataFrame({'New Address': Filter_Address})
New_HouseDS=pd.concat([HouseDS,New_Address],axis=1)
```

```python
New_HouseDS.head()
```

```python
New_HouseDS["Road Type"].value_counts()
```

```python
New_HouseDS.describe()
```

```python
#splitting Data set into Training and testing sets
X=New_HouseDS.drop(['Price','Title','Address','Area','Road','Road Width','Road Type','Amenities'],axis=1)
y=New_HouseDS['Price']
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
'''train_data=X_train.join(y_train)
train_data.hist(figsize=(15,8))'''
```

```python
print(X_train.shape,X_test.shape,y_train.shape,y_test.shape)
```

```python
#OneHotEncoding to convert String value to Numeric value
from sklearn.preprocessing import StandardScaler,OneHotEncoder
from sklearn.compose import ColumnTransformer
features=["City","New Address"]
one_hot=OneHotEncoder()
transformer = ColumnTransformer([("one_hot",one_hot,features)],remainder="passthrough")

transformed_X=transformer.fit_transform(X)
```

```python
pd.DataFrame(transformed_X)
```

```python
dummies=pd.get_dummies(New_HouseDS[["City","New Address"]])
dummies
```

```python
#feature scaling(to normalize feature variables)
np.random.seed(42)
scaler = StandardScaler(with_mean=False)
X_train,X_test,y_train,y_test=train_test_split(transformed_X,y,test_size=0.2,random_state=1)
X_train_scale=scaler.fit_transform(X_train)
X_test_scale=scaler.transform(X_test)
X_test_scale
```

```python
svm_model=SVR(kernel='linear')
train_data=X_train.join(y_train)
train_data.hist(figsize=(15,8))
```

```python
#Training the model
svm_model.fit(X_train_scale,y_train)
```

```python
#Making predictions on test set

pred=svm_model.predict(X_test_scale)
```

```python
#Evaluating the model
from sklearn.metrics import mean_squared_error,r2_score
mse=mean_squared_error(y_test,pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, pred)
```

```python
print("Mean Squared Error (MSE):", mse)
print("Root Mean Squared Error (RMSE):", rmse)
print("R-squared (R2):", r2)
```
